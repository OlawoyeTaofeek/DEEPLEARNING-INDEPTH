````python
def create_user(username: str, email: str, age: int):
    if not isinstance(username, str):
        raise TypeError("Username must be a string")
    if not isinstance(email, str):
        raise TypeError("Email must be a string")
    if not isinstance(age, int):
        raise TypeError("Age must be an integer")
    return {
        "username": username,
        "email": email,
        "age": age
    }

user1 = create_user("john_doe", "john@example.com", 30)
print(user1)
````

````python
from pydantic import BaseModel

class User(BaseModel):
    username: str
    email: str
    age: int

# Gives us type int and also validate the data
user2 = User(username="jane_doe", email="jane@example.com", age=25)
print(user2)
````

````python
type(user1)  # <class 'dict'>
type(user2)  # <class '__main__.User'>
````

````python
user3 = User(username="invalid_user", email=None, age="not_an_int")  # This will raise a validation error
````

````python
from pydantic import BaseModel, ValidationError
from datetime import datetime

class User(BaseModel):
    username: str
    email: str
    age: int
    signup_date: datetime

user = User(username="john_doe", email="john@example.com", age=30, signup_date=datetime.now())
print(user)
````

````python
from pydantic import BaseModel, ValidationError
from datetime import datetime
from typing import Union

class User(BaseModel):
    uid: int
    username: str
    email: str
    age: int
    verified_at: datetime | None = None
    bio: str = ""
    is_active: bool = True
    full_name: Union[str, None] = None

user = User(uid=123, username="john_doe", email="john@example.com", age=30, verified_at=datetime.now())
print(user)
````
